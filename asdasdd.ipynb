```python
import pandas as pd

df = pd.read_csv('./qweqwe.csv')
df
```

```python
df = df.drop(columns=['Unnamed: 0']).reset_index(drop=True)
```

```python
df
```

```python
profanity_list = ['ㅅㅂ','ㅆㅂ', 'ㅈㄴ', '존나', '씨발', '병신', 'ㅂㅅ', 'ㅄ', 'fuck', '뒤졌', '뒤졋', '디졋', 'ㄷㅈ', '디짐', '노꼴', 'ㅈ나', '씹', '개년', '개새끼', '개새', '개자식', '시발', '지랄', 'ㅈㄹ', '남창', '느금마', '니미럴', '애미', '애비', 'ㄴㄷㅆ', '니기미', '등신', '시팔', '씨팔']
```

```python
# 욕설이 포함된 행을 제거하는 함수를 정의합니다.
def remove_profanity_rows(df, profanity_list):
    # 욕설이 포함된 행을 필터링합니다.
    profanity_filter = df['Comment'].str.contains('|'.join(profanity_list), na=False)
    # 욕설이 포함되지 않은 행만 선택합니다.
    clean_df = df[~profanity_filter]
    return clean_df

# 욕설이 포함된 행을 제거합니다.
clean_df = remove_profanity_rows(df, profanity_list)

print(clean_df)
```

```python
clean_df = clean_df.reset_index(drop=True)
clean_df
```

```python
clean_df['Target'] = 0
```

```python
clean_df.to_csv('qweqwe.csv', index=False, encoding='utf-8-sig')
```
