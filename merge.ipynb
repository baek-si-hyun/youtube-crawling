```python
import pandas as pd

co_df = pd.read_csv('./merge_comments.csv')

bw_df = pd.read_csv('./merge_comments2.csv')
bw_df


```

```python
bw_df.columns = ['Comment', 'Target']
bw_df
```

```python
co_df.columns = ['Comment', 'Target']
co_df
```

```python
df_combined = pd.concat([co_df, bw_df], ignore_index=True)
df_combined
```

```python
df_combined.drop_duplicates(inplace=True)
```

```python
df_combined.duplicated().sum()
```

```python
display(df_combined[df_combined['Comment'].isna()])
df_combined = df_combined.dropna(subset=['Comment'])
```

```python
df_combined.Target.value_counts()
```

```python
# 타겟 데이터 언더샘플링
zero = df_combined[df_combined.Target == 0].sample(8225, random_state=124)
one = df_combined[df_combined.Target == 1]

df = pd.concat([zero, one])
df.reset_index(drop=True, inplace=True)
df
```

```python
df.Target.value_counts()
```

```python
df.to_csv('merge_comments_data.csv', index=False, encoding='utf-8-sig')
```
